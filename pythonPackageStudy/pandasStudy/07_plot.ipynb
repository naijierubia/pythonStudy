# pandas可视化

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ts = pd.Series(np.random.randn(10), index=pd.date_range("2020-01-01", periods=10))
ts
```

## 折线图

```python
ts.plot(grid=True)
```

```python
# 同时绘制多个图像
df = pd.DataFrame(np.random.randn(10, 2), columns=['B', 'C'])
df['A'] = np.arange(10)
df.plot(x='A', y=['B', 'C'])
```

## 柱状图

```python
apples = np.random.normal(50, 10, 12)   # 12个月 每个月苹果的销量
s = pd.Series(apples, index=pd.date_range('2019-01-01', periods=12, freq='M'))
s
```

```python
s.plot.bar()
```

```python
oranges = np.random.normal(50, 10, 12)  # 12个月 每个月橘子的销量
df = pd.DataFrame(
    {"apples": apples, "oranges": oranges},
    index=pd.date_range("2019-01-01", periods=12, freq="M"),
)
df
```

```python
df.plot.barh()
```

## 直方图

```python
data = np.random.randn(1000)
s = pd.Series(data)
s.plot.hist(bins=100)
```

## 散点图

```python
df.plot.scatter(x='apples', y='oranges', c='apples', colormap='jet', s=80)
```

## 饼状图

```python
data = np.array([22, 21, 25, 8])
labels = np.array(["Python", "JS", "Java", "C++"])
data = pd.Series(data, index=labels)
data.plot.pie(figsize=(4, 4))
```

```python
data = pd.DataFrame(np.random.normal(100, 10, (4, 4)), index=labels)
data.plot.pie(subplots=True, figsize=(8, 8), layout=(2, 2))
```

## 箱线图

```python
s = pd.Series(np.random.randn(100))
s.plot.box()
```

```python
df = pd.DataFrame({'A':pd.Series(np.random.randn(100)), 'B':pd.Series(np.random.randn(100)), 'C':pd.Series(np.random.randn(100))})
df.plot.box()
```
